## Preproccess data

```python
import pandas as pd
```

```python
df_business = pd.read_csv('/Users/erwinminor/Desktop/AI/business.csv')
```

```python
df_tip = pd.read_csv('/Users/erwinminor/Desktop/AI/tip.csv')
```

```python
df_tip['text'] = df_tip['text'].fillna('').astype(str)
df_tip = df_tip.groupby('business_id')['text'].agg(' '.join).reset_index()
```

```python
df_review = pd.read_csv('/Users/erwinminor/Desktop/AI/review.csv')
```

```python
df_review['text'] = df_review['text'].fillna('').astype(str)
df_review = df_review.groupby('business_id').agg({
    'text': ' '.join,
    'stars': 'mean'
}).reset_index()
```

```python
df_final = pd.merge(df_business, df_tip, on='business_id', how='left')
df_final = pd.merge(df_final, df_review, on='business_id', how='left')
```

```python
df_final.to_csv('/Users/erwinminor/Desktop/AI/final_data.csv')
```
